```python
import math
x=range(0,10)
y=[math.exp(-xi - 1) for xi in x]  # isto se chama "list comprehension"
print(x)
print(y)

import matplotlib.pyplot as plt
import numpy as np
plt.bar(x,y,color="red",align="edge",width=1)
x1 = np.linspace(0,10, 1000) # intervalo de 0 a 10, com 1000 divis√µes
y1=[math.exp(-xi) for xi in x1]
plt.plot(x1,y1)
plt.show()
```

```python
x_i = 0
x_f = 10
dx = 1
x = np.arange(x_i,x_f,dx)
S = 0
for xi in x:
    S = S+ math.exp(-xi - 1)*dx # *dx que √© 1 nesse caso
    
print("Soma de 10 caixinhas= ",S)
```
